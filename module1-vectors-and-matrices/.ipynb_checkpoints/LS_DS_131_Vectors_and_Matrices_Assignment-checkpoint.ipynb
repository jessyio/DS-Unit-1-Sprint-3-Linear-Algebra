## 1.1 Create a two-dimensional vector and plot it on a graph

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

two_d = [0.2,0.5]

plt.arrow(0,0,two_d[0],two_d[1], head_width=.02, head_length=0.01, color='y')
```

## 1.2 Create a three-dimensional vector and plot it on a graph

```python
from mpl_toolkits.mplot3d import Axes3D

three_d_1 = [0.1, 0.2, 0.3]
three_d_2 = [0.2, 0.3, 0.4]
three_d_3 = [0.3, 0.4, 0.5]


vectors = np.array([[0.5, 0, 0, .1, .2, .3], 
                    [0, 0, 0, .2, .3, .4],
                    [0, 0, 0, .3, .4, .5]])

#unpacking vectors into 6 arrays
x, y, z, u, v, w = zip(*vectors)
#x, y, z, u, v, w

#create the 3D graph
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

#plot the arrays on the graph
ax.quiver(x, y, z, u, v, w, length=1)

#setting where the arrays start to plot at 0,0,0
ax.set_xlim([0,1])
ax.set_ylim([0,1])
ax.set_zlim([0,1])

#giving the name of the axis
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```

## 2.2 Find the following products: CD, AE, and BA. What are the dimensions of the resulting matrices? How does that relate to the dimensions of their factor matrices?

```python
##matrix multiplication

#CD
#c: 2, 3; d: 3, 3

c = np.array([[9,6,3],
            [4,7,11]])

d = np.array([[1,0,0],
             [0,1,0],
             [0,0,1]])

np.matmul(c, d)

#should have a end matrix size of 6 (2*3)
#end matrix would be the same "c" as "d" is an identity matrix
```

```python
#AE
#a: 3,2; e: 2,2

a = np.array([[1,2],
             [3,4],
             [5,6]])

e = np.array([[1,3],
             [5,7]])

np.matmul(a, e)

#should have a end matrix size of 6 (3*2)
```

```python
#BA
#b: 1,3; a: 3,2

b = np.array([[2,4,6]])

np.matmul(b, a)

#should have a end matrix size of 2 (1*2)
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```python
#determinant
h = np.array([[12,11],
             [7,10]])

np.linalg.det(h)
```

```python
j = np.array([[0,1,2],
             [7,10,4],
             [3,2,0]])
np.linalg.det(j)
```
